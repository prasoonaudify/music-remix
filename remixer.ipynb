```
import torch
import librosa
import matplotlib.pyplot as plt
import tensorflow as tf
```

```
from nnAudio import Spectrogram
from torchvision.utils import save_image

path_to_audio_file = '/content/drive/MyDrive/music_remixer/audio_files/metal_00000.wav'
path_to_spectrograms = '/content/drive/MyDrive/music_remixer/spectrograms/'
signal, sr = librosa.load(path_to_audio_file)
signal_torch = torch.from_numpy(signal)

device = 'cuda' if torch.cuda.is_available() else 'cpu'

cqt_layer = Spectrogram.CQT1992v2(sr=sr, bins_per_octave = 36, verbose=False,
                                  trainable=False, output_format='Magnitude')

mel_layer = Spectrogram.MelSpectrogram(sr=sr, n_mels=80, n_fft=1024,
                                       hop_length=256, window='hann',
                                       pad_mode='constant', fmin=0.0, fmax=None)

cqt_spectrogram = cqt_layer(signal_torch)
mel_spectrogram = mel_layer(signal_torch)

save_image(cqt_spectrogram, path_to_spectrograms + 'cqt_spectrogram.png')
save_image(mel_spectrogram, path_to_spectrograms + 'mel_spectrogram.png')
```

```
from huggingface_hub import from_pretrained_keras
model = from_pretrained_keras("keras-io/MelGAN-spectrogram-inversion")
```

```
audio_sample = tf.random.uniform([128, 50, 80])
model.predict(audio_sample, batch_size = 32, verbose = 1)
```
