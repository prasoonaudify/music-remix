# GENRE CHANGE

```python
TEXT_PROMPT = ['metal']
AUDIO_FILE = '/content/drive/MyDrive/music-generation/audiocraft/music_files/shape_of_you.mp3'
model = MusicGen.get_pretrained('melody')
model.set_generation_params(duration=40)
melody, sr = torchaudio.load(AUDIO_FILE)
wav = model.generate_with_chroma(TEXT_PROMPT, melody[None].expand(len(TEXT_PROMPT), -1, -1), sr)
for idx, one_wav in enumerate(wav):
    audio_write(path_to_save + AUDIO_FILE.split('/')[-1].split('.')[0] + '-' + TEXT_PROMPT[0], one_wav.cpu(), model.sample_rate, strategy="loudness", loudness_compressor=True)
```

```python
IPython.display.display(IPython.display.Audio(path_to_save + AUDIO_FILE.split('/')[-1].split('.')[0] + '-' + TEXT_PROMPT[0]+'.wav'))
```

# BASS (heavy low frequency)


```python
def normalize(s):
    return s.lower().strip().replace("'", "").replace("(", "").replace(")","").replace(' ', '-')

TEXT_PROMPT = ['heavy low frequency']
AUDIO_FILE = '/content/drive/MyDrive/music-generation/audiocraft/music_files/levitating.mp3'
model = MusicGen.get_pretrained('melody')
model.set_generation_params(duration=40)
melody, sr = torchaudio.load(AUDIO_FILE)
wav = model.generate_with_chroma(TEXT_PROMPT, melody[None].expand(len(TEXT_PROMPT), -1, -1), sr)
for idx, one_wav in enumerate(wav):
    audio_write(path_to_save + AUDIO_FILE.split('/')[-1].split('.')[0] + '-' + normalize(TEXT_PROMPT[0]), one_wav.cpu(), model.sample_rate, strategy="loudness", loudness_compressor=True)
```

```python
IPython.display.display(IPython.display.Audio(path_to_save + AUDIO_FILE.split('/')[-1].split('.')[0] + '-' + normalize(TEXT_PROMPT[0])+'.wav'))
```
